```python
import pandas as pd
```

```python
# Load the dataset 
df_tweet = pd.read_csv('Tweet.csv')

df_tweet
```

```python
# Load the dataset 
df_company_tweet = pd.read_csv('Company_Tweet.csv')

df_company_tweet
```

```python
# Merge them
df_merged = pd.merge(df_tweet, df_company_tweet, on = 'tweet_id', how = 'inner')
df_merged.head()
```

```python
# Create date
df_merged['post_date'] = pd.to_datetime(df_merged['post_date'], unit='s')
df_merged['date'] = pd.to_datetime(df_merged['post_date'].apply(lambda date: date.date()))
df_merged.head()
```

```python
df_merged.drop(['tweet_id', 'writer', 'post_date'], axis = 1, inplace = True)
df_merged
```

```python
df_merged_1 = df_merged.iloc[:int((len(df_merged) / 10)),:]
df_merged_1
```

```python
df_merged_2 = df_merged.iloc[int((len(df_merged) / 10)):int((len(df_merged) / 10) * 2),:]
df_merged_2
```

```python
df_merged_3 = df_merged.iloc[int((len(df_merged) / 10) * 2):int((len(df_merged) / 10) * 3),:]
df_merged_3
```

```python
df_merged_4 = df_merged.iloc[int((len(df_merged) / 10) * 3):int((len(df_merged) / 10) * 4),:]
df_merged_4
```

```python
df_merged_5 = df_merged.iloc[int((len(df_merged) / 10) * 4):int((len(df_merged) / 10) * 5),:]
df_merged_5
```

```python
df_merged_6 = df_merged.iloc[int((len(df_merged) / 10) * 5):int((len(df_merged) / 10) * 6),:]
df_merged_6
```

```python
df_merged_7 = df_merged.iloc[int((len(df_merged) / 10) * 6):int((len(df_merged) / 10) * 7),:]
df_merged_7
```

```python
df_merged_8 = df_merged.iloc[int((len(df_merged) / 10) * 7):int((len(df_merged) / 10) * 8),:]
df_merged_8
```

```python
df_merged_9 = df_merged.iloc[int((len(df_merged) / 10) * 8):int((len(df_merged) / 10) * 9),:]
df_merged_9
```

```python
df_merged_10 = df_merged.iloc[int((len(df_merged) / 10) * 9):,:]
df_merged_10
```

```python
df_merged_1.to_csv('KAG_TWEETS_DF_1.csv', index=False)
df_merged_2.to_csv('KAG_TWEETS_DF_2.csv', index=False)
df_merged_3.to_csv('KAG_TWEETS_DF_3.csv', index=False)
df_merged_4.to_csv('KAG_TWEETS_DF_4.csv', index=False)
df_merged_5.to_csv('KAG_TWEETS_DF_5.csv', index=False)
df_merged_6.to_csv('KAG_TWEETS_DF_6.csv', index=False)
df_merged_7.to_csv('KAG_TWEETS_DF_7.csv', index=False)
df_merged_8.to_csv('KAG_TWEETS_DF_8.csv', index=False)
df_merged_9.to_csv('KAG_TWEETS_DF_9.csv', index=False)
df_merged_10.to_csv('KAG_TWEETS_DF_10.csv', index=False)
```
