A szükséges csomagok import-álása

```
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import time
```

URL generálásra használt függvény definiálása

```
def create_archive_url(page_number):
  return 'https://www.economist.com/business/?page=' + str(page_number)
```

Az Economist oldalán 857 oldalt lehet gyűjteni a Business témában, ezért a for ciklus 0-tól 845-ig megy. Csak a hírek címének és a dátumát mentem el. A try-catch arra szolgál, hogyha valamiért az oldal nem töltődne be, és hibát dobna, akkor ezt jelzem és utána azt külön letöltöm majd.
Itt ellenőríztem a lementett adatok méretét is.

```
title, date, error = [], [], []
df = pd.DataFrame()

for page_number in range(857):
  try:
    url = create_archive_url(page_number)

    html = requests.get(url).text
    page = bs(html, "html.parser")

    for i in page.find('div', class_='layout-section-collection ds-layout-grid').find_all('span', class_='teaser__headline teaser__headline--sc3'):
        title.append(i.text)

    for i in page.find('div', class_='layout-section-collection ds-layout-grid').find_all('a', class_='headline-link'):
        date.append(i['href'][10:20])

    time.sleep(3)

    print(page_number)

  except:
    error.append(page_number)

    print(str(page_number) + str(" error"))

print(title)
print(len(title))
print(date)
print(len(date))
```

Az elsőre nem sikerült hírek lekérése

```
print(len(error))
```

Az összegyűjtött adatok eltárolása DataFrame-ben.

```
df['title'] = title
df['date'] = date

df
```
