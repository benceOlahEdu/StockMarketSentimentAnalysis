A szükséges csomagok import-álása

```
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import time
```

URL generálásra használt függvény definiálása

```
def create_archive_url(page_number):
  return 'https://www.economist.com/finance-and-economics/?page=' + str(page_number)
```

Az Economist oldalán 845 oldalt lehet gyűjteni a Fincance témában, ezért a for ciklus 0-tól 845-ig megy. Csak a hírek címének és a dátumát mentem el. A try-catch arra szolgál, hogyha valamiért az oldal nem töltődne be, és hibát dobna, akkor egy hosszabb alvás után újra lekérem az oldalt, és akkor a hiba előfordulásának a valószínűsége nagyságrendekkel kisebb.

```
title, date = [], []
df = pd.DataFrame()

for page_number in range(845):
  try:
    url = create_archive_url(page_number)

    html = requests.get(url).text
    page = bs(html, "html.parser")

    for i in page.find('div', class_='layout-section-collection ds-layout-grid').find_all('span', class_='teaser__headline teaser__headline--sc3'):
        title.append(i.text)

    for i in page.find('div', class_='layout-section-collection ds-layout-grid').find_all('a', class_='headline-link'):
        date.append(i['href'][23:33])

    time.sleep(3)

  except:
    time.sleep(5)

    url = create_archive_url(page_number)

    html = requests.get(url).text
    page = bs(html, "html.parser")

    for i in page.find('div', class_='layout-section-collection ds-layout-grid').find_all('span', class_='teaser__headline teaser__headline--sc3'):
        title.append(i.text)

    for i in page.find('div', class_='layout-section-collection ds-layout-grid').find_all('a', class_='headline-link'):
        date.append(i['href'][23:33])

    time.sleep(3)

  print(page_number)

print(title)
print(len(title))
print(date)
print(len(date))
```

Az összegyűjtött adatok eltárolása DataFrame-ben.

```
df['title'] = title
df['date'] = date

df
```
